```python
# Memilih hanya kolom numerik
df_num = df_indonesia.select_dtypes(include=[np.number])

# Menghitung Q1 (persentil ke-25) dan Q3 (persentil ke-75)
Q1 = df_num.quantile(0.25)
Q3 = df_num.quantile(0.75)
IQR = Q3 - Q1

# Menentukan outlier jika data di luar rentang [Q1 - 1.5 * IQR, Q3 + 1.5 * IQR]
outliers = (df_num < (Q1 - 1.5 * IQR)) | (df_num > (Q3 + 1.5 * IQR))

# Menampilkan kolom dan baris yang mengandung outlier
outliers_info = df_num[outliers.any(axis=1)]
outliers_info
```

```python
# Creating a column 'region'

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name

df['region'] = df['country'].apply(country_to_continent)
df.head(10)
```

```python
from sklearn.preprocessing import scale

# standarisasi belum ditransformasi
standar_transform = pd.DataFrame(scale(df_sdg), columns=df_sdg.columns, index=df_sdg.index)

standar_transform
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
import geopandas as gpd

# K-Means Clustering
silhouette_scores = []
for n_clusters in range(2, 11):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    cluster_labels = kmeans.fit_predict(standar_transform)
    silhouette_avg = silhouette_score(standar_transform, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    
plt.plot(range(2, 11), silhouette_scores, marker="o")
plt.xlabel('Nomor Cluster')
plt.ylabel('Silhouette Score')
plt.title('Silhouette method')
plt.show()
```

```python
# menemukan centroid dan menentukan data ke cluster
kmeans = KMeans(n_clusters=6, n_init=10, random_state=42)
cluster_labels = kmeans.fit_predict(standar_transform)

cluster_centers= pd.DataFrame(kmeans.cluster_centers_, columns=df_sdg.columns)
cluster_centers
```

```python
def plot_cluster_map(cluster_labels, aggregated_df):
    # Load the world map data
    world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

    # Reset index to ensure it's unique and use country names as index
    aggregated_df = aggregated_df.reset_index()
    aggregated_df['Country'] = aggregated_df['country']  # Assuming 'index' column has country names

    # Ensure no duplicate country names
    aggregated_df = aggregated_df.drop_duplicates(subset=['Country'])

    # Add cluster labels to DataFrame
    df_with_clusters = aggregated_df.copy()
    df_with_clusters['Cluster'] = cluster_labels

    # Merge the world map data with the DataFrame containing cluster labels
    world_with_clusters = pd.merge(world, df_with_clusters, left_on='name', right_on='Country', how='left')

    # Plot the world map with cluster labels
    fig, ax = plt.subplots(figsize=(25, 7))
    world_with_clusters.plot(column='Cluster', cmap='viridis', linewidth=0.1, ax=ax, edgecolor='0.8', legend=True)
    plt.title('Clustering Map of the World')
    plt.show()
    
    return world_with_clusters

# Example usage
world_with_clusters = plot_cluster_map(cluster_labels, standar_transform)
```

```python
def add_cluster_column(aggregated_df, cluster_labels):
    # Ensure the index is reset and unique
    aggregated_df = aggregated_df.reset_index()

    # Add cluster labels as a new column to the DataFrame
    aggregated_df['Cluster'] = cluster_labels

    return aggregated_df

# Example usage:
# Assuming 'aggregated_df' is your DataFrame and 'cluster_labels' are the labels generated from KMeans
df_countries_cluster = add_cluster_column(aggregated_df, cluster_labels)

# Display the updated DataFrame
df_countries_cluster
```

```python
def plot_cluster(df_countries_cluster):
    cluster_counts = df_countries_cluster['Cluster'].value_counts().sort_index()
    
    plt.figure(figsize=(19, 6))
    bars = cluster_counts.plot(kind='bar', color='green')
    plt.xlabel('Cluster')
    plt.ylabel('Jumlah Negara')
    plt.title('Jumlah Negara Setiap Cluster')
    plt.xticks(rotation=0)
    
    for i, bar in enumerate(bars.patches):
        height = bar.get_height()
        bars.annotate(f"{height}", (bar.get_x() + bar.get_width() / 2, height),
                      ha='center', va='bottom')
    plt.show()
    
plot_cluster(df_countries_cluster)
```

```python
def sdg_index_comparison(top_countries_idn):
    sns.set(style='whitegrid')
    
    # Daftar warna khusus (hijau, kuning, merah, dan lain-lain)
    colors = ['green', 'yellow', 'brown', 'blue', 'orange', 'red']

    for score, aggregated_df in top_countries_idn.items():
        plt.figure(figsize=(12, 6))
        
        # Membuat barplot dengan warna yang telah ditentukan
        barplot = sns.barplot(x='country', y=score, data=aggregated_df, palette=colors[:len(aggregated_df)])
        
        plt.title(f'Perbandingan {score.replace("_", " ").title()}')
        plt.xlabel('Negara')
        plt.ylabel(score.replace("_", " ").title())
        
        # Menambahkan anotasi pada setiap bar
        for p in barplot.patches:
            barplot.annotate(format(p.get_height(), '.2f'),
                             (p.get_x() + p.get_width() / 2, p.get_height()),
                             ha='center', va='center',
                             xytext=(0, 9),
                             textcoords='offset points')   
        
        # Mengatur batas y-axis dan rotasi label x-axis
        plt.ylim(0, aggregated_df[score].max() * 1.1)
        plt.xticks(rotation=45)
        
        plt.show()

# Memanggil fungsi untuk menampilkan plot
sdg_index_comparison(top_countries_idn)
```
